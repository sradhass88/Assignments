```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
df = pd.read_csv('/content/Students_Performance.csv')
df.head(10)
```

a. Find out how many males and females participated in the test.

```python
count_gender = df['gender'].value_counts()
print(count_gender)
df['gender'].value_counts().plot(kind='bar',color = 'yellow',figsize=(8,4))
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Geneder Distribution')
plt.show()
```

b. What do you think about the students' parental level of education?

```python
# Display the count of each parental level of education
parental_education_counts = df['parental level of education'].value_counts()
print(parental_education_counts)
plt.figure(figsize=(10, 6),facecolor='white')
sns.countplot(y='parental level of education', data=df, order=df['parental level of education'].value_counts().index,color='red')
plt.title('Parental Level of Education Distribution')
plt.show()
```

c. Who scores the most on average for math, reading and
writing based on
i. Gender

```python
# Who Scores the Most on Average for Math, Reading, and Writing
# i. Based on Gender
average_scores_gender = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
print(average_scores_gender)
average_scores_gender.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()
```

 ii. Test preparation course

```python
# ii. Based on Test Preparation Course
average_scores_preparation = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
print(average_scores_preparation)
average_scores_preparation.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Test Preparation Course')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()
```

What do you think about the scoring variation for math, reading and writing based on i. Gender

```python
# Analyze the Scoring Variation for Math, Reading, and Writing
# i. Based on Gender
std_scores_gender = df.groupby('gender')[['math score', 'reading score', 'writing score']].std()
print(std_scores_gender)
std_scores_gender.plot(kind='bar', figsize=(10, 6))
plt.title('Score Variation by Gender')
plt.ylabel('Standard Deviation')
plt.xticks(rotation=45)
plt.show()
```

 ii. Test preparation course

```python
# Calculate the standard deviation of scores based on test preparation course completion
std_scores_preparation = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].std()
print(std_scores_preparation)

# Plot the standard deviation of scores by test preparation course
std_scores_preparation.plot(kind='bar', figsize=(10, 6))
plt.title('Score Variation by Test Preparation Course')
plt.ylabel('Standard Deviation')
plt.xticks(rotation=0)
plt.show()
```

e. The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this.

```python
# e. Identify the Top 25% of Students Based on Their Math Score
math_75th_percentile = df['math score'].quantile(0.75)
print(f"75th Percentile of Math Scores: {math_75th_percentile}")
```

```python
top_25_percent_math = df[df['math score'] >= math_75th_percentile]
print(top_25_percent_math)
```

```python
# Add bonus points to the top 25% of students
# Assuming we add 5 bonus points
bonus_points = 5
df['bonus points on math']=0
df.loc[df['math score'] >= math_75th_percentile, 'bonus points on math']= bonus_points
df
```

```python
# Display the updated scores
df['Updated Math score'] =df['math score']+df['bonus points on math']
df
```

```python
# Plot the distribution of math scores before and after adding bonus points
plt.figure(figsize=(10, 6))

# Original scores
plt.hist(df['math score'], bins=20, alpha=0.5, label='Original Math Scores', color='gray')

# Updated scores
plt.hist(df['Updated Math score'], bins=20, alpha=0.7, label='Updated Math Scores with Bonus', color='blue')

# Add titles and labels
plt.axvline(math_75th_percentile, color='red', linestyle='dashed', linewidth=2, label='75th Percentile')
plt.title('Distribution of Math Scores Before and After Adding Bonus Points')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.legend()

# Show the plot
plt.show()
```

# Case Study on Testing of Hypothesis

**2.** A company started to invest in digital marketing as a new
way of their product promotions. For that they collected
data and decided to carry out a study on it.
a. The company wishes to clarify whether there is any
increase in sales after stepping into digital marketing.
b. The company needs to check whether there is any
dependency between the features â€œRegionâ€ and
â€œManagerâ€.
Help the company to carry out their study with the help of data
provided.


```python
import pandas as pd
from scipy import stats

df = pd.read_csv('/content/Sales_add.csv')
df.head(10)
```

```python
# a. The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.
t_stat, p_value = stats.ttest_rel(df['Sales_before_digital_add(in $)'], df['Sales_After_digital_add(in $)'], alternative='greater')

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant increase in sales after digital marketing.")
else:
    print("Fail to reject the null hypothesis: There is no significant increase in sales after digital marketing.")
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Region'], df['Manager'])

# Perform chi-square test of independence
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
alpha =0.05
#H0:There is no dependency between Region and Manager.
#H1:There is a dependency between Region and Manager.

# Determine if the result is significant
if p_value < alpha:
    print("Reject the null hypothesis: There is a dependency between Region and Manager.")
else:
    print("Fail to reject the null hypothesis: There is no dependency between Region and Manager.")
```
